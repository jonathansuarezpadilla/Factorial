```python
%%writefile config.py
import numpy as np

# === Parámetros de entrada ===
produccion_deseada = 19584 # kg/día
PM_SO3 = 80.06 # kg/kmol
T = 707.9 # K
P0_atm = 5.5 # atm
R = 0.082057 # atm·m³/kmol·K

# Constantes para conversión
R_m3 = 8.314 # J/mol·K = Pa·m³/mol·K
P_Pa = 101325 # 1 atm en Pa
N_avogadro = 6.02214076e23 # moléculas/mol

# Geometría y catalizador
rho_b = 256 # [kg/m³] densidad del lecho
D_reactor = 0.31 # [m]
Ac = np.pi * (D_reactor / 2) ** 2 # [m²]
dp = 0.005961 # diametro de particula [m]
phi = 0.303 # porosidad del lecho
mu = 3.43e-5 # [Pa·s] viscosidad del flujo de gas de alimentación
include_pressure_drop = False
conversion_objetivo = 0.95
W_span = (0.0, 0.5) # Rango de masa de catalizador [kg]

# Composición de la alimentación (fracciones molares)
y_A_feed = 0.07 # SO2 (A)
y_B_feed = 0.13 # O2 (B)
y_F_feed = 0.79 # N2 (F)
y_D_feed = 0.01 # H2O (D)

# Masa molar species (kg/kmol)
M_A, M_B, M_C, M_D, M_E, M_F, M_G = 64.06, 32.00, 80.06, 18.02, 98.08, 28.013, 30.01

# === Flujos molares (inlet) ===
produccion_kmol_dia = produccion_deseada / PM_SO3 # [kmol/dia]
F_C_produccion = produccion_kmol_dia / (24 * 3600) # [kmol/s]
F_A_in = F_C_produccion / conversion_objetivo # [kmol/s]
F_total_in = F_A_in / y_A_feed
F_B_in = F_total_in * y_B_feed
F_F_in = F_total_in * y_F_feed
F_C_in = 0.0 # No hay SO3 a la entrada
F_G_in = 0.0 # No hay NO a la entrada
F_D_in = F_total_in * y_D_feed
F_E_in = 0.0 # No hay H2SO4 a la entrada

# Condiciones iniciales para la ODE: [F_A, F_C, F_B, F_G, F_F, F_D, F_E, P]
y0 = [F_A_in, F_C_in, F_B_in, F_G_in, F_F_in, F_D_in, F_E_in, P0_atm]
```

```python
%%writefile kinetics.py
import numpy as np
from config import T, rho_b, R_m3, P_Pa, N_avogadro

# === Constantes Cinéticas (Reacción 1: A + 1/2 B <-> C) ===
def calcular_constantes_cineticas(T):
    """Calcula las constantes de equilibrio (Kp) y velocidad (k1, K2, K3) para la Reacción 1."""
  Kp = np.exp((11300 / T) - 10.68)
  k1 = np.exp(12.16 - (5473 / T))
  K2 = np.exp(-9.953 + (8619 / T))
  K3 = np.exp(-71.745 + (52596 / T))
  return Kp, k1, K2, K3

# === Tasa de Reacción 2: C(SO₃) + D(H₂O) → E(H₂SO₄) ===
def calcular_tasa_reaccion_2(T, C_C, C_D, rho_b):
    """
    Calcula la tasa de la reacción 2 en kmol/(kg_cat·s).
    C_C, C_D deben estar en mol/m³.
    """
    # Constante cinética en moléculas/cm³·s
    k_l_molecules = 2.26e-43 * T * np.exp(6544 / T) # cm³/(molécula²·s)

    # Conversión de mol/m³ a moléculas/cm³
    factor_conversion = N_avogadro / 1e6
    C_C_moleculas = C_C * factor_conversion
    C_D_moleculas = C_D * factor_conversion

    # Tasa en moléculas/(cm³·s)
    r_2_molecules = k_l_molecules * C_C_moleculas * (C_D_moleculas ** 2)

    # Conversión a kmol/(m³·s)
    factor_conversion_2 = 1000 * 1e6 / N_avogadro
    r_2_kmol_m3s = r_2_molecules * factor_conversion_2

    # Conversión a kmol/(kg_cat·s)
    r_2 = r_2_kmol_m3s / rho_b

    return r_2

# === Tasa de Reacción 3: F(N₂) + B(O₂) → 2G(NO) ===
def calcular_tasa_NO(T, P_B, P_F, P_G, rho_b):
    """
    Calcula la tasa de formación de NO (r_3) en kmol/(kg_cat·s).
    P_B, P_F, P_G: Presiones parciales (atm).
    """
    # Concentraciones en mol/m³
    C_B = (P_B * P_Pa) / (R_m3 * T)
    C_F = (P_F * P_Pa) / (R_m3 * T)
    C_G = (P_G * P_Pa) / (R_m3 * T)

    # Constantes de velocidad (ANSYS, 2009)
    kf1 = 1.8e8 * np.exp(-38370 / T) # m³/mol·s
    kf2 = 1.8e4 * np.exp(-4680 / T) # m³/mol·s
    kr1 = 3.8e7 * np.exp(-425 / T) # m³/mol·s
    kr2 = 3.81e4 * np.exp(-20820 / T) # m³/mol·s
    kO = 3.81e5 * (T ** -0.5) * np.exp(-31090 / T) # mol⁰·⁵/m¹·⁵·s

    # Evitar divisiones por cero
    C_B = max(C_B, 1e-20)
    C_F = max(C_F, 1e-20)

    # Tasa en mol/m³·s
    numerador = 1 - (kr1 * kr2 * C_G**2) / (kf1 * C_F * kf2 * C_B)
    denominador = 1 + (kr1 * C_G) / (kf2 * C_B)

    r_3_mol_m3s = 2 * kf1 * kO * (C_B ** 0.5) * C_F * (numerador / denominador)

    # Convertir a kmol/(kg_cat·s)
    r_3 = r_3_mol_m3s / (rho_b * 1000.0)

    return r_3
```

```python
%%writefile reactor.py
import numpy as np

# Importar todo lo necesario
from config import (
    T, rho_b, Ac, include_pressure_drop, R, R_m3, P_Pa, M_A, M_B, M_C, M_D, M_E, M_F, M_G,
    dp, phi, mu
)
from kinetics import calcular_constantes_cineticas, calcular_tasa_reaccion_2, calcular_tasa_NO

# Constantes cinéticas calculadas una sola vez
Kp, k1, K2, K3 = calcular_constantes_cineticas(T)

# Estados: [F_A, F_C, F_B, F_G, F_F, F_D, F_E, P] (F_i en kmol/s, P en atm)
def pbr_ode(W, Y):

    # === Variables de estado ===
    F_A, F_C, F_B, F_G, F_F, F_D, F_E, P = Y

    # Control de mínimos para evitar errores numéricos
    P = max(P, 1e-9)
    # Se asegura que los flujos no sean cercanos a cero para los cálculos de presiones parciales
    F_A_m, F_C_m, F_B_m, F_G_m, F_F_m, F_D_m = np.maximum([F_A, F_C, F_B, F_G, F_F, F_D], 1e-20)
    F_E_m = max(F_E, 1e-20)

    FT = F_A_m + F_C_m + F_B_m + F_G_m + F_F_m + F_D_m + F_E_m
    if FT <= 0:
         FT = 1e-12

    # Presiones parciales (atm)
    p_A = P * F_A_m / FT
    p_C = P * F_C_m / FT
    p_B = P * F_B_m / FT
    p_G = P * F_G_m / FT
    p_F = P * F_F_m / FT
    p_D = P * F_D_m / FT
    p_E = P * F_E_m / FT

    # === Reacción 1: r1: A(SO₂) + 1/2 B(O₂) → C(SO₃) ===
    if p_A > 0 and p_B > 0:
        beta = p_C / (Kp * p_A * np.sqrt(max(p_B, 1e-12)))
    else:
        beta = 0.0
    term_rev = max(0.0, 1.0 - beta)

    numer = k1 * p_B * p_A * term_rev
    denom = 22.414 * (1.0 + K2 * p_A + K3 * p_C) ** 2
    r_1 = numer / denom # kmol/(kg·s)

    # === Reacción 2: C(SO₃) + D(H₂O) → E(H₂SO₄) ===
    C_C = (p_C * P_Pa) / (R_m3 * T) # mol/m³
    C_D = (p_D * P_Pa) / (R_m3 * T) # mol/m³
    r_2 = calcular_tasa_reaccion_2(T, C_C, C_D, rho_b) # kmol/(kg_cat·s)

    # === Reacción 3: F(N₂) + B(O₂) → 2G(NO) ===
    r_3 = calcular_tasa_NO(T, p_B, p_F, p_G, rho_b) # kmol/(kg_cat·s)

    # === BALANCES MOLARES: dFi/dW ===
    # Reacción 1: A + 1/2 B -> C
    # Reacción 2: C + D -> E
    # Reacción 3: F + B -> 2G

    dF_A_dW = -r_1
    dF_C_dW = +r_1 - r_2
    dF_B_dW = -0.5 * r_1 - r_3
    dF_G_dW = +2 * r_3
    dF_F_dW = -r_3
    dF_D_dW = -r_2
    dF_E_dW = +r_2

    # === Caída de presión (Ergun) ===
    if include_pressure_drop:
        # Densidad de mezcla (kg/m³)
        M_mix = (F_A * M_A + F_C * M_C + F_B * M_B + F_G * M_G + F_F * M_F + F_D * M_D + F_E * M_E) / FT
        rho_g = (P * M_mix) / (R * T) # R en atm·m³/kmol·K

        # Flujo volumétrico (m³/s) y velocidad superficial (m/s)
        v = FT * R * T / P
        u = v / Ac

        term1 = (150.0 * (1.0 - phi) ** 2 * mu * u) / (phi ** 3 * dp ** 2)
        term2 = (1.75 * (1.0 - phi) * rho_g * u ** 2) / (phi ** 3 * dp)
        dPdz_Pa = - (term1 + term2)
        dPdz_atm = dPdz_Pa / 101325.0
        dPdW = dPdz_atm / (rho_b * Ac)
    else:
        dPdW = 0.0

    return [dF_A_dW, dF_C_dW, dF_B_dW, dF_G_dW, dF_F_dW, dF_D_dW, dF_E_dW, dPdW]
```

```python
%%writefile calculos.py
import numpy as np

def calcular_selectividades_y_rendimientos_objetivo(F_A_in, F_A_obj, F_C_obj, F_E_obj, F_G_obj):
    """
    Calcula las selectividades y rendimientos en el punto objetivo (W_req).

    Args:
        F_A_in (float): Flujo molar de A (SO2) de entrada.
        F_A_obj (float): Flujo molar de A (SO2) en el punto objetivo.
        F_C_obj (float): Flujo molar de C (SO3) en el punto objetivo.
        F_E_obj (float): Flujo molar de E (H2SO4) en el punto objetivo.
        F_G_obj (float): Flujo molar de G (NO) en el punto objetivo.

    Returns:
        dict: Diccionario con todas las selectividades y rendimientos calculados.
    """
    results = {}

    # Selectividad del SO3 (deseado) vs H2SO4 + NO (no deseados)
    denom_total_no_deseado = F_E_obj + F_G_obj
    results['selectividad_SO3_H2SO4_NO'] = F_C_obj / denom_total_no_deseado if denom_total_no_deseado > 1e-12 else 0.0

    # Selectividad SO3 vs H2SO4
    results['selectividad_SO3_H2SO4'] = F_C_obj / F_E_obj if F_E_obj > 1e-12 else 0.0

    # Selectividad SO3 vs NO
    results['selectividad_SO3_NO'] = F_C_obj / F_G_obj if F_G_obj > 1e-12 else 0.0

    # Rendimiento (Yield basado en reactivo consumido)
    denom_reactivo_consumido = F_A_in - F_A_obj
    results['rendimiento_SO3_consumido'] = F_C_obj / denom_reactivo_consumido if denom_reactivo_consumido > 1e-12 else 0.0

    # Rendimiento (Yield basado en reactivo alimentado - Yield/Rendimiento alternativo)
    results['yield_SO3_alimentado'] = F_C_obj / F_A_in

    return results


def calcular_perfiles_selectividad_y_rendimiento(W, F_A_in, F_A, F_C, F_E, F_G):
    """
    Calcula los arrays de selectividad, rendimiento y yields a lo largo del reactor.

    Args:
        W (np.array): Array de masa de catalizador.
        F_A_in (float): Flujo molar de A (SO2) de entrada.
        F_A, F_C, F_E, F_G (np.array): Arrays de flujos molares a lo largo de W.

    Returns:
        dict: Diccionario con los arrays de resultados para graficar.
    """
    N = len(W)
    results = {}

    # 1. Arrays de Selectividad
    selectividad_total_array = np.zeros(N)
    selectividad_H2SO4_array = np.zeros(N)
    selectividad_NO_array = np.zeros(N)

    for i in range(N):
        # Selectividad total
        denom_total_no_deseado = F_E[i] + F_G[i]
        if denom_total_no_deseado > 1e-12:
            selectividad_total_array[i] = F_C[i] / denom_total_no_deseado

        # Selectividad vs H2SO4
        if F_E[i] > 1e-12:
            selectividad_H2SO4_array[i] = F_C[i] / F_E[i]

        # Selectividad vs NO
        if F_G[i] > 1e-12:
            selectividad_NO_array[i] = F_C[i] / F_G[i]

    results['selectividad_total_array'] = selectividad_total_array
    results['selectividad_H2SO4_array'] = selectividad_H2SO4_array
    results['selectividad_NO_array'] = selectividad_NO_array

    # 2. Array de Rendimiento (Yield basado en reactivo consumido)
    rendimiento_SO3_array = np.zeros(N)
    for i in range(N):
        denom_reactivo_consumido = F_A_in - F_A[i]
        if denom_reactivo_consumido > 1e-12:
            rendimiento_SO3_array[i] = F_C[i] / denom_reactivo_consumido

    results['rendimiento_SO3_array'] = rendimiento_SO3_array

    # 3. Arrays de Yields (basados en reactivo alimentado)
    results['yield_SO3_array'] = F_C / F_A_in
    results['yield_H2SO4_array'] = F_E / F_A_in
    results['yield_NO_array'] = F_G / F_A_in
    results['yield_total_array'] = results['yield_SO3_array'] + results['yield_H2SO4_array'] + results['yield_NO_array']

    return results

# Este módulo no debe ejecutarse directamente, solo importar sus funciones.
if __name__ == '__main__':
    print("Módulo 'calculos.py' cargado. Contiene funciones para selectividad y rendimiento.")
```

```python
%%writefile main.py
# main.py: Script principal para la simulación y graficado.
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.interpolate import interp1d # Necesario para interpolación precisa de W_req
from scipy.optimize import brentq # Necesario para encontrar la raíz (W_req)

# Los módulos config y reactor se importan directamente porque están en el mismo directorio (/content)
from config import (
    W_span, y0, F_A_in, P0_atm, T, rho_b, Ac, conversion_objetivo, F_total_in,
    F_B_in, F_F_in, F_D_in
)
from reactor import pbr_ode
# --- IMPORTACIÓN DE MÓDULO DE CÁLCULO ---
from calculos import calcular_selectividades_y_rendimientos_objetivo, calcular_perfiles_selectividad_y_rendimiento
# ----------------------------------------

def simular_pbr():
    """Ejecuta la simulación del reactor PBR y procesa los resultados."""

    # === Solución del ODE ===
    sol = solve_ivp(pbr_ode, W_span, y0, method='RK45', rtol=1e-7, atol=1e-10, dense_output=True)
    W = sol.t
    # Desempacar los resultados (F_A, F_C, F_B, F_G, F_F, F_D, F_E, P)
    F_A, F_C, F_B, F_G, F_F, F_D, F_E, P = sol.y

    # === Cálculo de FT y conversión ===
    FT = F_A + F_C + F_B + F_G + F_F + F_D + F_E
    X = 1.0 - F_A / F_A_in

    # === Encontrar W requerido y valores objetivo (Usando Interpolación para precisión) ===
    f_X = interp1d(W, X, kind='linear')

    try:
        # Encontrar W_req buscando la raíz donde X(W) - X_obj = 0
        W_req = brentq(lambda w: f_X(w) - conversion_objetivo, W[0], W[-1])

        # Obtener valores objetivo interpolados en W_req
        Y_obj = sol.sol(W_req)

        P_obj, X_obj, FT_obj = Y_obj[7], conversion_objetivo, np.sum(Y_obj[:7])
        F_A_obj, F_C_obj, F_B_obj, F_G_obj, F_F_obj, F_D_obj, F_E_obj = Y_obj[0], Y_obj[1], Y_obj[2], Y_obj[3], Y_obj[4], Y_obj[5], Y_obj[6]
        V_req = W_req / rho_b

    except ValueError:
        # Si la conversión objetivo no se alcanza
        W_req, V_req = np.nan, np.nan
        P_obj, X_obj, FT_obj = P[-1], X[-1], FT[-1]
        F_A_obj, F_C_obj, F_B_obj, F_G_obj, F_F_obj, F_D_obj, F_E_obj = F_A[-1], F_C[-1], F_B[-1], F_G[-1], F_F[-1], F_D[-1], F_E[-1]
        X_obj = X[-1] # Usar la conversión máxima alcanzada

    # =========================================================
    # === CÁLCULOS DE SELECTIVIDAD Y RENDIMIENTO ===
    # =========================================================

    # 1. Cálculos en el punto objetivo (W_req)
    if not np.isnan(W_req):
        results_obj = calcular_selectividades_y_rendimientos_objetivo(
            F_A_in, F_A_obj, F_C_obj, F_E_obj, F_G_obj
        )
        # Rendimientos (Yields) para impresión
        yield_SO3 = results_obj['yield_SO3_alimentado']
        yield_H2SO4 = F_E_obj / F_A_in
        yield_NO = F_G_obj / F_A_in
        yield_total = yield_SO3 + yield_H2SO4

        # Selectividades para impresión
        selectividad_SO3 = results_obj['rendimiento_SO3_consumido'] # Rendimiento SO3/SO2_reacted
        F_A_reacted = F_A_in - F_A_obj
        selectividad_H2SO4 = F_E_obj / F_A_reacted if F_A_reacted > 1e-9 else 0.0 # H2SO4/SO2_reacted
        selectividad_NO_vs_SO2_reacted = F_G_obj / F_A_reacted if F_A_reacted > 1e-9 else 0.0 # NO/SO2_reacted

        # Selectividad total SO3 / (H2SO4 + NO) para la gráfica combinada
        S_SO3_total_obj = results_obj['selectividad_SO3_H2SO4_NO']

    else:
        # Valores por defecto si no se alcanza la conversión
        results_obj = {}
        yield_SO3, yield_H2SO4, yield_NO, yield_total = 0.0, 0.0, 0.0, 0.0
        selectividad_H2SO4, selectividad_SO3, selectividad_NO_vs_SO2_reacted = 0.0, 0.0, 0.0
        S_SO3_total_obj = np.nan

    # 2. Cálculos a lo largo de todo el reactor (para graficar)
    perfiles_calc = calcular_perfiles_selectividad_y_rendimiento(
        W, F_A_in, F_A, F_C, F_E, F_G
    )

    # Re-asignar arrays de yield (basados en reactivo alimentado)
    yield_SO3_array = perfiles_calc['yield_SO3_array']
    yield_H2SO4_array = perfiles_calc['yield_H2SO4_array']
    yield_NO_array = perfiles_calc['yield_NO_array']
    yield_total_array = yield_SO3_array + yield_H2SO4_array

    # Perfil de Selectividad total SO3 / (H2SO4 + NO)
    S_SO3_total_array = perfiles_calc['selectividad_total_array']

    # --- Mostrar resultados (mantengo esta sección) ---
    print("=" * 60)
    print("RESULTADOS DEL REACTOR PBR - TRES REACCIONES")
    print("=" * 60)
    print(f"Condiciones de entrada:")
    print(f"  F_A(SO2)_in: {F_A_in:.6f} kmol/s")
    print(f"  F_B(O2)_in: {F_B_in:.6f} kmol/s")
    print(f"  F_F(N2)_in: {F_F_in:.6f} kmol/s")
    print(f"  F_D(H2O)_in: {F_D_in:.6f} kmol/s")
    print(f"  F_total_in: {F_total_in:.6f} kmol/s")
    print(f"  Presión: {P0_atm:.2f} atm")
    print(f"  Temperatura: {T:.1f} K")

    if not np.isnan(W_req):
        print(f"\nResultados en conversión objetivo ({conversion_objetivo*100:.1f}%):")
        print(f"  W requerido: {W_req:.4f} kg")
        print(f"  V requerido: {V_req:.6f} m³")
        print(f"  Longitud del reactor: {V_req/Ac:.2f} m")
        print(f"  Presión: {P_obj:.4f} atm")
        print(f"  Conversión alcanzada: {X_obj:.4f}")
        print(f"\nFlujos molares a W_obj (kmol/s):")
        print(f"  F_A(SO2): {F_A_obj:.6f}")
        print(f"  F_C(SO3): {F_C_obj:.6f}")
        print(f"  F_B(O2): {F_B_obj:.6f}")
        print(f"  F_F(N2): {F_F_obj:.6f}")
        print(f"  F_G(NO): {F_G_obj:.6f}")
        print(f"  F_D(H2O): {F_D_obj:.6f}")
        print(f"  F_E(H2SO4): {F_E_obj:.6f}")
        print(f"  Flujo total: {FT_obj:.6f}")

        # Selectividades
        print(f"\nSelectividades SO₃ vs Subproductos a W_obj:")
        print(f"  S(SO₃/H₂SO₄+NO): {S_SO3_total_obj:.6f}")
        print(f"  S(SO₃/H₂SO₄): {results_obj['selectividad_SO3_H2SO4']:.6f}")
        print(f"  S(SO₃/NO): {results_obj['selectividad_SO3_NO']:.6f}")

        # Rendimientos vs SO2 reaccionado
        print(f"\nRendimientos vs SO₂ reaccionado (Selectividades en algunos textos):")
        print(f"  Rend. SO₃ / SO₂-reac: {selectividad_SO3:.6f}")
        print(f"  Rend. H₂SO₄ / SO₂-reac: {selectividad_H2SO4:.6f}")
        print(f"  Rend. NO / SO₂-reac: {selectividad_NO_vs_SO2_reacted:.6f}")


    else:
        print(f"\nNo se alcanzó la conversión objetivo. Últimos valores:")
        print(f"  W máximo: {W[-1]:.4f} kg")
        print(f"  Conversión máxima: {X[-1]:.4f}")

    print(f"\nYields (Rendimientos vs SO₂ alimentado) a W_obj:")
    print(f"  Yield SO3/SO2: {yield_SO3:.6f} ({yield_SO3*100:.2f}%)")
    print(f"  Yield H2SO4/SO2: {yield_H2SO4:.6f} ({yield_H2SO4*100:.2f}%)")
    print(f"  Yield NO/SO2: {yield_NO:.6f} ({yield_NO*100:.2f}%)")
    print(f"  Yield total (SO3+H2SO4): {yield_total:.6f} ({yield_total*100:.2f}%)")
    print(f"  Verificación - Conversión: {X_obj:.6f}, Suma de Yields (todos): {yield_SO3 + yield_H2SO4 + yield_NO:.6f}")

    # =========================================================
    # === GRÁFICOS: 8 PLOTS EN UNA CUADRÍCULA 2x4 ===
    # =========================================================
    # Total de plots: 8 (4 de la Fig 1 original + 1 de Fig 2 + 1 de Fig 3 + 1 de Fig 4 + 1 de Fig 5)
    # Se organizarán en 2 filas y 4 columnas.

    plt.figure(figsize=(24, 10)) # Figura grande para 8 subplots (2 filas x 4 columnas)
    plt.suptitle('Perfil completo del reactor PBR (8 Gráficas en 2 Filas x 4 Columnas)', fontweight='bold', fontsize=16)

    # ------------------ FILA 1 (Posiciones 1 a 4) ------------------

    # Posición 1: Conversión de SO2 vs W (Original 1.1)
    plt.subplot(2, 4, 1)
    plt.plot(W, X, label='Conversión SO₂', color="blue")
    if not np.isnan(W_req):
        plt.axvline(W_req, linestyle='--', color='r', label=f'W_req = {W_req:.4f} kg')
        plt.axhline(conversion_objetivo, linestyle='--', color='g', label=f'X_obj = {conversion_objetivo:.2f}')
        plt.plot(W_req, X_obj, 'ro', markersize=8)
    plt.xlabel('W [kg]')
    plt.ylabel('Conversión X')
    plt.title('1. Conversión de SO₂')
    plt.grid()
    plt.legend(fontsize='small')

    # Posición 2: Caída de presión vs W (Original 1.2)
    plt.subplot(2, 4, 2)
    plt.plot(W, P, label='Presión Total', color="black")
    if not np.isnan(W_req):
        plt.axvline(W_req, linestyle='--', color='r')
        plt.plot(W_req, P_obj, 'ko', markersize=8)
    plt.xlabel('W [kg]')
    plt.ylabel('Presión [atm]')
    plt.title('2. Caída de presión')
    plt.grid()
    plt.legend(fontsize='small')

    # Posición 3: Flujos molares principales vs W (Original 1.3)
    plt.subplot(2, 4, 3)
    plt.plot(W, F_A, label='F_A(SO₂)', color='red')
    plt.plot(W, F_C, label='F_C(SO₃)', color='green')
    plt.plot(W, F_B, label='F_B(O₂)', color='blue')
    plt.plot(W, F_D, label='F_D(H₂O)', color='orange')
    plt.plot(W, F_E, label='F_E(H₂SO₄)', color='purple')
    plt.plot(W, F_G, label='F_G(NO)', color='cyan')

    if not np.isnan(W_req):
        plt.axvline(W_req, linestyle='--', color='k')
        plt.plot(W_req, F_A_obj, 'o', color='red', markersize=6)
        # Puntos objetivo para los demás flujos (F_C_obj, F_B_obj, etc.)
        # ... (Mantengo los puntos de la versión anterior)

    plt.xlabel('W [kg]')
    plt.ylabel('Flujo molar [kmol/s]')
    plt.title('3. Flujos molares (Todos)')
    plt.grid()
    plt.ylim(0, 0.006)
    plt.legend(loc='upper right', fontsize='x-small')

    # Posición 4: Formación de H₂SO₄ vs W (Original 1.4)
    plt.subplot(2, 4, 4)
    plt.plot(W, F_E, label='F_E(H₂SO₄)', color='purple')
    if not np.isnan(W_req):
        plt.axvline(W_req, linestyle='--', color='k')
        plt.plot(W_req, F_E_obj, 'o', color='purple', markersize=8)
    plt.xlabel('W [kg]')
    plt.ylabel('Flujo molar [kmol/s]')
    plt.title('4. Formación de H₂SO₄')
    plt.grid()
    plt.legend(fontsize='small')

    # ------------------ FILA 2 (Posiciones 5 a 8) ------------------

    # Posición 5: Formación de NO vs W (Original Figura 2)
    plt.subplot(2, 4, 5)
    plt.plot(W, F_G, label='F_G(NO)', color='cyan')
    if not np.isnan(W_req):
        plt.axvline(W_req, linestyle='--', color='k')
        plt.plot(W_req, F_G_obj, 'o', color='cyan', markersize=8)
    plt.xlabel('W [kg]')
    plt.ylabel('Flujo molar [kmol/s]')
    plt.title('5. Formación de NO')
    plt.grid()
    plt.legend(fontsize='small')

    # Posición 6: Yields y Conversión vs W (Original Figura 3)
    plt.subplot(2, 4, 6)
    plt.plot(W, yield_SO3_array, label='Yield SO₃ (C)', color='green', linewidth=2)
    plt.plot(W, yield_H2SO4_array, label='Yield H₂SO₄ (E)', color='purple', linewidth=2)
    plt.plot(W, yield_NO_array, label='Yield NO (G)', color='cyan', linewidth=2)
    plt.plot(W, X, label='Conversión SO₂ (A)', color='blue', linestyle=':',linewidth=2)
    plt.plot(W, yield_total_array, label='Yield Total (SO₃+H₂SO₄)', color='red', linestyle='--', linewidth=2)
    if not np.isnan(W_req):
        plt.axvline(W_req, linestyle='--', color='k', alpha=0.7)
        # Puntos objetivo (mantengo los puntos de la versión anterior)
    plt.xlabel('W [kg]')
    plt.ylabel('Yield / Conversión')
    plt.title('6. Yields y Conversión')
    plt.grid(True)
    plt.legend(loc='lower right', fontsize='x-small')

    # Posición 7: Selectividad vs W (Todas las Selectividades) (Original Figura 4)
    plt.subplot(2, 4, 7)
    plt.plot(W, S_SO3_total_array, label='S SO₃ / (H₂SO₄ + NO)', color='red', linewidth=2)
    plt.plot(W, perfiles_calc['selectividad_H2SO4_array'], label='S SO₃ / H₂SO₄', color='purple', linestyle='--', linewidth=2)
    plt.plot(W, perfiles_calc['selectividad_NO_array'], label='S SO₃ / NO', color='cyan', linestyle=':', linewidth=2)
    if not np.isnan(W_req) and 'selectividad_SO3_H2SO4_NO' in results_obj:
        plt.axvline(W_req, linestyle='--', color='k', alpha=0.7)
        # Puntos objetivo (mantengo los puntos de la versión anterior)
    plt.xlabel('W [kg]')
    plt.ylabel('Selectividad (S)')
    plt.title('7. Perfiles de Selectividad')
    plt.grid(True)
    plt.legend(loc='lower right', fontsize='x-small')

    # Posición 8: Selectividad y Rendimiento Combinados (Original Figura 5)
    plt.subplot(2, 4, 8)
    plt.plot(W, S_SO3_total_array, label='S SO₃/(H₂SO₄+NO)', color='red', linewidth=2)
    plt.plot(W, yield_SO3_array, label='Yield SO₃/SO₂', color='blue', linestyle='--', linewidth=2)
    if not np.isnan(W_req):
        plt.axvline(W_req, linestyle='--', color='k', alpha=0.7)
        plt.plot(W_req, S_SO3_total_obj, 'o', color='red', markersize=8)
        plt.plot(W_req, yield_SO3, 'o', color='blue', markersize=8)
    plt.xlabel('W [kg]')
    plt.ylabel('Selectividad / Rendimiento')
    plt.title('8. S y Yield (SO₃) Combinados')
    plt.grid(True)
    plt.legend(loc='lower right', fontsize='small')


    plt.tight_layout(rect=[0, 0, 1, 0.98]) # Ajusta el layout para que el título no se superponga
    plt.show()

simular_pbr()
```
